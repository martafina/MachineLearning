```python
import random
import numpy as np
import time

n = 100
m = 100
mat_a = [[random.randint(1, 1000) for e in range(n)] for e in range(m)]

mat_b = [[random.randint(1, 1000) for e in range(n)] for e in range(m)]

mat_c = [[random.randint(0, 0) for e in range(n)] for e in range(m)]
start = time.time()
for i in range(n):
    for j in range(len(mat_b[0])):
        for k in range(len(mat_b)):
            mat_c[i][j] += mat_a[i][k] * mat_b[k][i]
end = time.time()
```

```python
print("Matrix A: ")
for a in mat_a:
    print(a)
```

```python
print("\n Matrix B: ")
for b in mat_b:
    print(b)
```

```python
print("\n Result:")
for c in mat_c:
    print(c)
```

```python
print("\n Time of execution: ", end - start)
```

```python
print("\n\t\t Numpy")
a = np.array(mat_a)
b = np.array(mat_b)
print("\n Matrix A: ")
print(a)
print("\n Matrix B: ")
print(b)
print("\n Result: ")
start_np = time.time()
c = a.dot(b)
end_np = time.time()
print(c)
print("\n Time of execution: ", end_np - start_np)
```
